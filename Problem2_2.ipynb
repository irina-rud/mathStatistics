```python
import math
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
%matplotlib inline


tetta = 1
bounds = [-1000, 1000]
step = 0.01
new_range = np.arange(bounds[0], bounds[1] + step, step)

max_x = 0
maxF = float("-Inf")
cauchy_file = 'cauchy.txt'
X = []

f = open(cauchy_file) 
X = [float(line[:-1]) for line in f]

def log_likelihood_func(X, x0): #так как константа не влияет на точку, в которой достигается экстремум
                                #то забъём на константу
    global maxF
    global max_x
    
    res = 0.0
    for x in X:
        res -= math.log(1 + (x - x0) ** 2)
    if (res > maxF):
        maxF = res
        max_x = x0
    return res

def plot_likelihood(X):
    Func_arr = []
    Func_arr += [log_likelihood_func(X, x) for x in new_range]
    
    plt.plot(new_range, Func_arr ,'m', label = r'$ln_{p(X_{1}...X_{N})}(x) + N \ln \pi$')
    plt.title('  N = ' + str(len(X)))
    plt.xlabel(r'$x_{0}$')
    plt.legend(loc = 'best')
 
    plt.grid(True)
    plt.show()
    
plot_likelihood(X[0:round(len(X)/2):])
print("x_max = " + str(max_x))

max_x = 0
maxF = float("-Inf")
    
plot_likelihood(X)
print("x_max = " + str(max_x))
#экстремумы достигаются более-менее в одной точке.
```
