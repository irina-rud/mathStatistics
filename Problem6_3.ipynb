```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as sps
```

```python
N = 100
loc = 1
scale = 0
X = np.random.standard_cauchy(N)
```

Сопряжённое априорное распределение нормального с известной дисперсией $σ^2$:

$N(\mu_0,\, \sigma_0^2)$

$P(|a−\theta|<2\sigma)≈0.954>0.95$  по правилу 2$\sigma$

В качестве априорного распределения возьмём $N(0,\frac{1}{4^2})$, тогда $2\sigma=0.5.$

```python
variance = 1
mean = 0
def bayes_estimation(X, params):
    pmean,pvar = params
    return (pmean/pvar + np.sum(X)/variance)/(1/pvar + len(X)/variance)

param = [0,1/16]

def label(params):
    pmean,pvar = params
    return r'$\frac{\frac{' + str(pmean) + r'}{' + str(pvar) + r'}+\Sigma X_i}{\frac{1}{' + str(pvar) + r'}+n}$'

A_X = np.arange(1,N+1)
plt.figure(figsize=(20, 10))
B = [bayes_estimation(X[0:k+1],param) for k in range(len(X))]
plt.plot(A_X,np.abs(B),color='r',label=label(param))

M = [np.mean(X[0:k+1]) for k in range(N)]    
plt.plot(A_X,np.abs(np.add(M,-mean)), color='g', label=r'$\overline{X}$')
plt.legend(loc='lower left', prop={'size':30})

plt.xlim(20, 100)
plt.ylim(0, 3)
plt.show()
```
