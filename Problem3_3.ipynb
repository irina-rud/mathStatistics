```python
import math
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline


N= 10**3
step = 10**(-2)
K = 500
newRange = [i*step for i in range (10**2)]

def dispr(tetta):
    X = sps.bernoulli.rvs(p=tetta, size=N)
    summ = 0
    sum2 = 0
    thetta_zv = np.mean(X)
    X_boot = [sps.bernoulli.rvs(p=thetta_zv, size=N) for i in range(K)]
    for i in range (K):
        boot_thetta = np.mean(X_boot[i])
        summ += boot_thetta
        sum2 += boot_thetta ** 2
    return sum2 / K - (summ/K) ** 2

min_dispr = [thetta*(1 - thetta)/N for thetta in (newRange)]
boot_dispr = [dispr(theta) for theta in newRange]

plt.figure(figsize=(15, 5))  

plt.plot(newRange, min_dispr, color = 'c', label = r'$\frac{1}{I_{X} (\theta)}$')
plt.plot(newRange, boot_dispr, color = 'r', label = r'$bootstrap \quad D_{\theta}\theta ^ {*}}$')
legend = plt.legend(loc='upper center', shadow=True, fontsize='x-large')
legend.get_frame().set_facecolor('#00FFCC')


plt.legend(loc = (1.01,0))  
plt.show()
#минимально возможная дисперсия в точке 1/2 достигает своей наивольшей величины
```
